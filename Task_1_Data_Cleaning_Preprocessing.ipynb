```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset/marketing_campaign.csv", sep="\t")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Income"].median()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df["dt_customer"].head()
```

```python
df["dt_customer"] = pd.to_datetime(df["dt_customer"], format="%d-%m-%Y")
```

```python
df.info()
```

```python
print(df["education"].unique())
print(df["marital_status"].unique())
```

```python
df["education"] = df["education"].str.strip()
df["marital_status"] = df["marital_status"].str.strip()
```

```python
df.describe()
```

```python
df.to_csv("cleaned_dataset/marketing_campaign_cleaned.csv", index=False)
```

# Task 1: Data Cleaning and Preprocessing

## Summary of Changes

- Loaded the Customer Personality Analysis dataset using Pandas.
- Checked the dataset structure using `df.info()` and `df.head()`.
- Identified missing values using `df.isnull().sum()`.
- Filled 24 missing values in the `income` column using the median.
- Checked for duplicate records using `df.duplicated().sum()` and found no duplicates.
- Renamed all column names to lowercase for consistency.
- Converted the `dt_customer` column from string to datetime format.
- Standardized text columns by removing leading and trailing spaces.
- Reviewed the dataset statistics using `df.describe()` to identify potential outliers.
- Saved the cleaned dataset as `marketing_campaign_cleaned.csv`.
